```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([[640, 970, 1120, 1580, 1900, 2130, 2460, 2610, 2880, 3200]]).T
y = np.array([[120, 123, 182, 167, 240, 284, 251, 272, 335, 340]]).T
plt.scatter(x, y)
plt.xlabel('Size (Sq Ft)')
plt.ylabel('Price, £ (x 1000)')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
print(x_test)
print(y_test)
```

```python
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
# Train the model
regr.fit(x_train, y_train)
# Display the coefficients
print('Intercept: \n', regr.intercept_)
print('Coefficient: \n', regr.coef_)
```

```python
# Specify the mininmum and maximum values of x to display the line of best fit
x_range = np.array([min(x), max(x)])
# Now use the regression model to obtain the predictions for the lowest and highest x values
y_pred_range = regr.predict(x_range)
```

```python
# Plot the line along with the training data
plt.plot(x_range, y_pred_range, color='red')
# Plot the training data
plt.scatter(x_train, y_train)
# Plot the test data
plt.scatter(x_test, y_test,marker='x',color='black')
plt.xlabel('Size (Sq Ft)')
plt.ylabel('Price, £ (x 1000)')
plt.show()
```

```python
# Predicting the Test set results
y_pred = regr.predict(x_test)
print(y_test)
print(y_pred)
```

```python
from sklearn.metrics import r2_score
print('Coefficient of determination, R^2: %.3f'
 % r2_score(y_test, y_pred))
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error

rmse = np.sqrt(mean_squared_error(y_test,y_pred))
print(f'RMSE:{rmse:.3f}')
```

```python
mae = mean_absolute_error(y_test, y_pred)
print(f'MAE:{mae:.3f}')
```
